**LISTADO DE LOS 5 PRODUCTOS CON LAS MEJORES RESEÑAS**

```python
#Extraer "id_product" y "score" de la lista "lifestore_sales"
id_reviews_not_separated = [[sale[1], sale[2]] for sale in lifestore_sales]

#Extraer "id_product" y "name" de la lista "lifestore_products"
id_name_products = [[producto[0], producto[1]] for producto in lifestore_products]

#Extraer " id_product" y "score" de "lifestore_sales"
id_score = [[sale[1], sale[2]] for sale in lifestore_sales]
```

```python
#Crear diccionario en el que por cada producto guardaremos una lista de sus reviews
id_score_count = {}

for producto in id_score:
    id = producto[0]
    score = producto[1]
    if id not in id_score_count.keys():
        id_score_count[id] = []
    id_score_count[id].append(score)

#print(id_score_count)

#Convertir las llaves del diccionario "id_reviews_count" a una lista 
keys_id_lista = list(id_score_count.keys())
#print(keys_id_lista)
```

```python
#Crear lista de los promedios de cada producto con reseñas 

import math
from sympy import multiplicity
prom_products=[]
num_score = []
for id_product in id_score_count.keys():
    lista_score = id_score_count[id_product]
    promedio = sum(lista_score) / len(lista_score)
    
    num_scores = len(lista_score)
    num_score.append(num_scores)
    
    # Arreglo promedio a 2 decimales
    decimales = 2
    multiplicador = 10 ** decimales
    promedio = math.ceil(promedio * multiplicador) / multiplicador
    
    prom_products.append(promedio)

#print(prom_products)
#print(num_score)
```

```python
#Crear lista con los id y promedio de las reseñas de cada producto
id_score_numscore= []
i=0

for n in prom_products:
    lis=[keys_id_lista[i],prom_products[i], num_score[i]]
    id_score_numscore.append(lis)
    i += 1
#print(id_score_numscore)


#Ordenar las lista "id_score_numscore"
id_score_numscore = sorted(id_score_numscore, key=lambda x: x[1])
#print(id_score_numscore)
```

```python
#Obtener el id, reseña y numero de reseñas de los 5 productos con mejores reseñas 
top_id = []
reseña = []
num_reseña = []
for producto in id_score_numscore[-18:]:
        id = producto[0]
        reseñas = producto[1]
        num_reseñas= producto[2]
        top_id.append(id)
        reseña.append(reseñas)
        num_reseña.append(num_reseñas)
#print(top_id, reseña, num_reseña)
```

**TOTAL DE INGRESOS MENSUALES Y VENTAS PROMEDIO MENSUALES**

```python
# Ventas por meses 
id_fecha = [ [sale[0], sale[3]] for sale in lifestore_sales if sale[4] == 0 ]
#Diccionario para categorizar 
categorizacion_meses = {}

for par in id_fecha:
    id = par[0]
    _, mes, _ = par[1].split('/')
    
    if mes not in categorizacion_meses.keys():
        categorizacion_meses[mes] = []
    categorizacion_meses[mes].append(id)
```

```python
ventas= []
for key in categorizacion_meses.keys():
    lista_mes = categorizacion_meses[key]
    suma_venta = 0
    for id_venta in lista_mes:
        indice = id_venta - 1
        info_venta = lifestore_sales[indice]
        id_product = info_venta[1]
        precio = lifestore_products[id_product-1][2]
        suma_venta += precio
        
        
    print('En el mes: ',key, 'se vendidieron: $',suma_venta, f'-> Numero de ventas totales: {len(lista_mes)}')
    ventas.append(suma_venta)
```

**VENTAS TOTALES**

```python
ventas_totales = sum(ventas)
print('El ingreso anual es: $', ventas_totales)
```
